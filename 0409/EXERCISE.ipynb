```python
class Restaurant():
    def describe_restaurant(self):
        print("餐廳名稱:哈根達斯")
        print("烹飪的方式:泰國炒冰")
```

```python
flavor = ['草莓','香蕉','巧克力','抹茶']
```

```python
class IceCreamStand(Restaurant):
    def show_flavors(self):
        print('我喜歡的口味:', flavor)
```

```python
x=Restaurant()
```

```python
x.describe_restaurant()
```

```python
y=IceCreamStand()
```

```python
y.show_flavors()
```

```python
class A(object):
    def go(self):
        print ("go A go!")
    def stop(self):
        print ("stop A stop!")
    def pause(self):
        raise Exception("Not Implemented")
class B(A):
    def go(self):
        super(B, self).go()
        print ("go B go!")
class C(A):
    def go(self):
        super(C, self).go()
        print ("go C go!")
    def stop(self):
        super(C, self).stop()
        print ("stop C stop!")
class D(B,C):
    def go(self):
        super(D, self).go()
        print ("go D go!")
    def stop(self):
        super(D, self).stop()
        print ("stop D stop!")
    def pause(self):
        print ("wait D wait!")
class E(B,C):
    pass
```

```python
a = A()
b = B()
c = C()
d = D()
e = E()
```

```python
print('*')
a.go()
print('*')
b.go()
print('*')
c.go()
print('*')
d.go()
print('*')
e.go()
```

```python
print('*')
a.stop()
print('*')
b.stop()
print('*')
c.stop()
print('*')
d.stop()
print('*')
e.stop()
```

```python
a.pause()
b.pause()
c.pause()
d.pause()
e.pause()
```
